# Scores_linear_regression[다항회귀]

```python
import tensorflow as tf
import numpy as np
```

```python
xy = np.loadtxt('data-02-test-score.csv', delimiter=',',skiprows=1,
                dtype=np.float32)
xy.shape
```

```python
x_train = xy[:,:-1]
y_train = xy[:,[-1]]
print(x_train.shape, y_train.shape)
```

```python
xy
```

```python
# X * W = Y
# (m,n) * (n,l) = (m,l)
#(25,3) * (3,1) = (25,1)
W = tf.Variable(tf.random.normal([3,1]),name='weight')
b = tf.Variable(tf.random.normal([1]),name='bias')
print(W)
print(b)
```

```python
# 예측함수 : H(X1, X2, X3) = W1*X1+ W2*X2+ W3*X3 + b
def hypo(X):
    return tf.matmul(X,W) + b # 내적 곱셈
```

```python
# 비용 함수 : (Hx - y)^2의 평균
# tf.square()
def cost_func():
    cost = tf.reduce_mean(tf.square(hypo(x_train) - y_train))
    return cost
```

```python
optimizer = tf.keras.optimizers.Adam(learning_rate=0.01)
```

```python
# 학습 시작
print('***** Start Learning!!')
for step in range(10000):
    with tf.GradientTape() as tape:
        cost = cost_func()
    gradients = tape.gradient(cost, [W, b])
    optimizer.apply_gradients(zip(gradients, [W, b]))
    
    if step % 100 == 0:
        print('%04d' % step, '손실 값: [', cost.numpy(), ']',
              ' W:', W.numpy(), ' b:', b.numpy())
print('***** Learning Finished!!')
```

```python
# 예측
test = [[ 73.,  80.,  75.],
       [ 93.,  88.,  93.],
       [ 89.,  91.,  90.],
       [ 96.,  98., 100.],
       [ 73.,  66.,  70.]]
test_1 = np.array(test,dtype=np.float32)
pred = hypo(test_1)
print(pred)
```

```python
# 정확도 측정 : RMSE
def get_rmse(y_test,preds):
    squared_error = 0
    for k,_ in enumerate(y_test):
        squared_error += (preds[k] - y_test[k])**2
    mse = squared_error/len(y_test)  
    rmse = np.sqrt(mse)
    return rmse[0]

x_test = x_train
y_test = y_train

preds = hypo(x_test).numpy()
print('Rmse:',get_rmse(y_test,preds))
```
